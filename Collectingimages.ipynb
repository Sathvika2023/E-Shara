```python
import cv2
import os
import time
import uuid
```

```python
# Define your labels and number of images per label
labels = ['hello', 'thanks', 'yes', 'no', 'how are you']
number_images = 15
# IMAGES_PATH = r'C:\Users\Sathvika\RealTimeObjectDetection\Tensorflow\workspace\images\collectedimages'
#  # Update to the actual path of your created folder

# # Check if the base directory exists
# if not os.path.exists(IMAGES_PATH):
#     print(f"Error: Base directory '{IMAGES_PATH}' does not exist.")
# else:
#     print(f"Using base directory: {IMAGES_PATH}")

# Define the relative path for images
IMAGES_PATH = os.path.join(os.getcwd(), 'collectedimages')

# Ensure the base directory exists or create it
os.makedirs(IMAGES_PATH, exist_ok=True)

# Then, within the code, use this IMAGES_PATH to save images.

# Loop through each label
for label in labels:
    # Create a directory for each label inside the base path
    label_path = os.path.join(IMAGES_PATH, label)
    os.makedirs(label_path, exist_ok=True)  # Creates subdirectory for each label
    print(f"Directory created (or already exists) for label '{label}' at: {label_path}")
    
    # Start video capture
    cap = cv2.VideoCapture(0)
    print(f'Collecting images for {label}')
    time.sleep(2)  # Allow some time to prepare for image collection
    
    # Capture the specified number of images
    for imgnum in range(number_images):
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture image")
            continue

        # Generate a unique filename for each image
        imagename = os.path.join(label_path, f"{label}_{str(uuid.uuid1())}.jpg")
        cv2.imwrite(imagename, frame)
        cv2.imshow('frame', frame)
        
        print(f"Captured image {imgnum + 1} for label '{label}' at {imagename}")
        time.sleep(2)  # Delay between captures
        
        # Exit loop if 'q' key is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    # Release the video capture and close the display window for each label
    cap.release()
    cv2.destroyAllWindows()

print("Image collection completed.")
```
